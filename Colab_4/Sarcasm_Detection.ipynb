```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
```

```python
file_path = '/content/drive/MyDrive/data_crawl.csv'
df = pd.read_csv(file_path)
```

```python
df = shuffle(df)
df.reset_index(inplace = True)
df.drop(['Unnamed: 0', 'index'], inplace=True, axis=1)
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df
```

```python
df_copy = df.copy()
df2 = df_copy[['headline', 'is_sarcastic']]
```

```python
df2
```

```python
def lower_function(x):
    x = x.lower()
    return x

import re
def del_num_function(x):
  re.sub('[^A-Za-z0-9 ]', '', x)
  return x
```

```python
df2['headline'] = df2['headline'].apply(lower_function)
df2['headline'] = df2['headline'].apply(del_num_function)
```

```python
df2
```

```python
paragraph = ""
for i in df2['headline']:
  paragraph = paragraph + i + " "
word_in_headline = paragraph.split()
```

```python
# Kiểm tra tần suất từ 
from nltk import FreqDist
frequency = FreqDist(word_in_headline)
frequency.most_common(20)
```

```python
# Tokenization
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

headline_token = []
for i in df2['headline']:
  headline_token.append(word_tokenize(i))
```

```python
headline_token[:2]
```

```python
# Delete Stop Word
from nltk.corpus import stopwords
stop_word = stopwords.words('english')

headline_token_delSW = []
for i in headline_token:
  k = []
  for j in i:
    if (j not in stop_word):
      k.append(j)
  headline_token_delSW.append(k)
```

```python
# lemmatization/stemming
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
  
wnl = WordNetLemmatizer()

headline_token_delSW_lemma = []
for i in headline_token_delSW:
  k = []
  for j in i:
    root = wnl.lemmatize(j)
    k.append(root)
  headline_token_delSW_lemma.append(k)
```

```python
for i in headline_token_delSW_lemma[:5]:
  print(i)
```

```python
data = []
for i in headline_token_delSW_lemma:
  k = " ".join(i)
  data.append(k)

label = df2['is_sarcastic']
```

```python
data[:10]
```

```python
data_dict = {'headline': data, 'is_sarcastic': label}
```

```python
df_final = pd.DataFrame(data_dict)
```

```python
df_final
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, label_train, label_test = train_test_split(df_final['headline'], df_final['is_sarcastic'], train_size = 0.7, shuffle=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

transformed_data_train = vectorizer.fit_transform(data_train.values).toarray()
vocab_in_data_train = vectorizer.vocabulary_

vectorizer = CountVectorizer(vocabulary = vocab_in_data_train)
transformed_data_test = vectorizer.fit_transform(data_test.values).toarray()
```

```python
print(vectorizer.get_feature_names())
print(len(vectorizer.get_feature_names()))

print(transformed_data_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

model = MultinomialNB(alpha=1.1)
model.fit(transformed_data_train, label_train)
label_pred = model.predict(transformed_data_test)
```
